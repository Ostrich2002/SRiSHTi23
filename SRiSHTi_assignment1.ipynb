```python
#Q2 function

def highest_paid_female_worker(work_hours):
    max_hours = 0
    max_worker = None

    for worker in work_hours:
        name, hours, gender = worker
        if gender == 'F' and hours > max_hours:
            max_hours = hours
            max_worker = name

    if max_worker is not None:
        payment = max_hours * 20
        print(f"The female worker who worked the most hours is {max_worker}.")
        print(f"She should be paid ${payment}.")
    else:
        print("No female workers found.")

work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'), ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]
highest_paid_female_worker(work_hours)
```
